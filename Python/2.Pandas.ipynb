### Filtering

Selecting/Extracting the data based on the condition(s) 

Using 1 condition 
Syntax: df[(condition)]

```python
b = df[df['Gender'] =='Male']
b.head()
```

```python
df[(df['tip']>=7)]
```

```python
df[~(df['tip']>=7)]
```

```python
df[(df['tip']<7)]
```

Using multiple conditions

Syntax: df[(condition1) or/and (condition2)]

```python
df[(df['tip']>=7) & (df['Gender']=='Male')]
```

```python
df[(df['tip']>3) | (df['Gender']=='Female')]
```

```python
df[(df['Gender']== 'Female') | (df['tip']>=3) & (df['total_bill']<=60)]
```
