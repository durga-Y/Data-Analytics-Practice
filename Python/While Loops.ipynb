#### While loop 

* repeates the block of code until the condition satisfies
* to iterate a block of code as long as the condition is true 

```python
x=1
while x<=5:
    print('ds')
    x=x+1
    
```

### While loop with else 

* The else part is executed if the condition in the while loop evaluates to false ,The while loop can be terminated with break statement
* in such case the else part is ognored hence a while loop's else part runs if no break uccur and the condition is false
  

```python
i=1
while i<=3:
    print(i)
    i=i+1
    if i==1:
        break

else :
    print('d')
    
```
