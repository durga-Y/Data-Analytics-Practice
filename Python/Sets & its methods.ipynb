#### Set Methods 

##### Set.add() --------->This methods adds an element  to the set, if element already exists in the set it will simply ignore return nothing  

```python
s = {1,2,3,4,}
s.add(5)
print(s)
```

##### Set.update() ---------> This method updates the set with another set or any iterable 

```python
s1 = {1,2,3}
s2 = {1,2,3,4,5}
s1.update(['x', 'y','z'])
print(s1)
```

##### Set.Discard() --------> This method removes the specified element from the set if elememnt does not exist in the set it willl ignire (nothing returns 

```python
s = {1,2,3,4,5}
s.discard(3)
print(s)
```

```python
s.discard(7)
print(s)          # nothing will happen 
```

##### Set.remove() -------> This method removes the specified element from the set if element does not exist in the set it will through the error

```python
s= {1,2,3}
s.remove(2)
print(s)
```

```python
s.remove(7)
print(s)        # will through the error 
```

##### Set.clear() -----> This methos will remove all the elements from the set 

```python
s.clear()
print(s)
```

```python
del s # Deletes the complete variable 
print(s)
```

''' Set.copy() ------> This method returns the shallow copy of set 
    Changes to the copied set doesnt affect the original set ''' 

```python
s1 = {1,2,3,}
s2=s1.copy()
s2.add(4)
print(s1)
print(s2)
```

##### Set.union() ----> This method returns a new set with elemts from both sets

```python
A= {2,3,4,}
B = {4,5,6               # removes duplicates 
A.union(B)
```

##### Set.intersection()-----> This method returns a new set with elements that are common to all sets (returns only which are common )


```python
A= {2,3,4,5}
B = {4,5,6,3}
A.intersection(B)
```

##### Set.isdisjoint() ---> Returns True if two sets have no intersection else returns fals 

```python
s1={1,2,3}
s2={4,5,6}
s3={7,5,9}
s=s1.isdisjoint(s2)
print(s)
s=s2.isdisjoint(s3)
print(s)


```

##### Set.symmetric_difference()-----> This mothod returns new set with elments that are either of the sets but not in both(not commonelements)
##### (A union B) - (A intersection B)

```python
s1={1,2,3}
s2={2,3,4}
s1.symmetric_difference(s2)
```

##### Set.difference()-------> This method returns a new set with elements in the sets that are in the other specified sets  

```python
s1={1,2,3}
s2={2,3,4}
print(s1-s2)
print(s2-s1)
```

##### Set.issubset()---------> method returns True if all elements of first set are available in second set else returns False

```python
s1={3,4,5}
s2={3,4,5,6}
A=s1.issubset(s2)
print(A)
A=s2.issubset(s1)
print(A)
```

##### Set.issupersset()----->method returns true if all elements of second set are available in first set else return False

```python
s1={3,4,5}
s2={3,4,5,6}
A=s1.issuperset(s2)
print(A)
A=s2.issuperset(s1)
print(A)
```
