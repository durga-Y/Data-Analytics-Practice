#### Pandas Series 

it is 1 dimensional (1-D)array holding data as any time 

A pandas series is like a column in a table  

pandas series can be created using list 

```python
l = [10,20,30]
s = pd.Series(l)
print(s)
```

```python
type(s)
```

### Pandas

```python
df = pd.DataFrame({'Age': [26,38,41,32,39],'Salary':[28000,42000,60000,55000,50000],'Gender':['M','F','F','M','F']})
print(df)
                  
```

Set index

```python
d = df.set_index('Age')
print(d)
```

Reset index

```python
df.reset_index()
print(d)
```

Drop a single row 

```python
df = df.drop(index=[3])
print(df)
```

drop multiple rows

```python
print(df)
```

```python
df1 = df.drop(index=[0,1,2])
print(df1)
```

Drop a single column 

```python
df1 = df.drop(columns=['Age'])
print(df1)
              
```

Drop multiple columns 

```python
df1 = df.drop(columns=['Age','Gender'])
print(df1)
              
```

Sorting values in ascending order 

```python
df = df.sort_values(by='Age', ascending = True)
print(df)
```

```python
df= df.sort_values(by='Salary', ascending=True)
print(df)
```

Sorting values in Descending order 

```python
df = df.sort_values(by='Age', ascending = False)
print(df)
```

### Conmbining Data Frames

```python
df1 = pd.DataFrame({'city':['mumbai','delhi','banglore','hyderabad'],'temperature':[32,45,40,36]})
df1
```

```python
df2 = pd.DataFrame({'city':['mumbai','delhi','banglore','chennai'],'humidity':[68,65,75,70]})
df2
```

#### Concatenation

```python
pd.concat([df1,df2], axis=1) 
```

```python
pd.concat([df1,df2], axis=0) 
```

```python
pd.concat([df1,df2], ignore_index=True) 
```

#### Merging of DataFrame

```python
# Left join
df_left = pd.merge(df1,df2, on = 'city', how = 'left')
df_left
```

```python
#right join
df_right = pd.merge(df1,df2, on='city', how='right')
df_right
```

```python
#inner joine ---> common records 
df_inner=pd.merge(df1,df2, on='city', how='inner')
df_inner
```

```python
# outer join --->all records of all data frames 
df_outer=pd.merge(df1,df2, on='city',how='outer')
df_outer
```
