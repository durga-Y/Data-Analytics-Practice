#### Block Structure 

* The code that is executed when a specific specific condition is met is defined in a 'block'
* statements preceding blocks generally ends with a colon (:)
* in python the block structure is signalled by changes in indentation
* Each line of code in certain block level must be indented equally and indented  more than the surrounding scope
* the standed is to use 4 spaces for each level of block indentation

### If- elif- else 

```python
x = 10 
if x<0:
    print('positive')

elif x==0:
    print('Zero')

else:
    print('negative')
```

### Nested if statements 

We can have a if...elif...else statement inside another if...elif...else statement this is called nesting in computer programming 

```python
num = 10
if num > 0:
    if num == 0:
        print("Zero")
    else:
        print("positive number")

else :
    print("negative number")
    
```
