### Funtions for creating 1-D array

#### Arrange
* arrange is an array-valued version of the built-in python range function
* Syntax: np.arange(start,end,step)  

```python
a=np.arange(10)
print(a)
```

#### Linspace

* return evenly spaces numbers over a specified interval.
* Syntax: np.linspace(start,end,num_of_samples)

```python
np.linspace(1,5,5)
```

##### Sort

```python
# sorting along an axis: axis=1--->row, axis=0----->column
import numpy as np
a = np.array([[2,5,3],[7,4,3],[9,4,3]])
b = np.sort(a,axis=0)
print(b)
```

```python
import numpy as np
a = np.array([[1,2],[3,4],[5,6]])
b = np.array([[9,8],[7,6],[5,4]])
# adding a and b
print(a+b)
```

```python
# dot product--> multiply a and b 9elemntwise multiplcation )
a*b
```

```python
a = np.array([[1,2],[3,4]])
b = np.array([[9,8],[7,6]])
# matrix multiplcation 
np.matmul(a, b)
```

```python
np.concatenate((a,b), axis=0)
```

```python
#vertical stack         # equal to np.concatenate(a,b),axis=0
np.vstack((a,b))
```

```python
np.concatenate((a,b), axis=1)
```

```python
#horizontal stack         # equal to np.concatenate(a,b),axis=1
np.hstack((a,b))
```
