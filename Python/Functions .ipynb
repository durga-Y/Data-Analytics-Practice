#### 5. Arbitary Arguments

* Sometimes we do not no in advance athe number of arguments that will be passed into the finction, python aloows us to handle this kind of situation with function calls with arbitary number of arguments.

```python
#defining the function 
def greet(*n):
    for i in n:
        print("Hello", i)
```

```python
#Call the function
greet(1,2,3,4)
```

#### Recursive funstion or recursion 
* A function can call other function, even it is possible for the function to call itself.these type of construct is termed as recursive function  

```python
#define the function
def factorial(x):
    if type(x)==int:
        if x==0:
            return 1
        elif x>=0:
            return x*factorial(x-1)
        else:
            print("value error")

#call the function
factorial(5)
            
    
```

#### Anonymous (Lambda) functions
* primarily used to temporarly define a finctiion for use by another function.

```python
#define a function the "usual" way
def squared(x):
    return x**2

#call the function 
squared(5)
```

```python
# define a identical function using lambda
squared = lambda x: x**2
#call function
squared(2)
```
