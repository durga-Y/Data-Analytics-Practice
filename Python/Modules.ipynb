#### Random module 

* This module defines several functions to generate random numbers
* we can use these functions while developing games, in cryptography, and to generate random numbers on fly for authentication.

```python
#random() - funtion: will always genrate a some random float value between 0 and 1 
import random 
random.random()
```

```python
#randint() - function: to genarate random ingegers between to given values inclusive of end value)
random.randint(1, 100)
```

```python
#randrange() - finction: to genrate random integer or integer in given range
random.randrange(2,50,2)
```

```python
#Choice() - function: it wont return random number,it will retun a random object from the given list 
List = ["bunny","chinny","minny","vinny"]
random.choice(List)
```
