### Measures of dispertion or measures of spread (Second business momemnt )

Range , IQR, Variance, std.deviation

```python
df =pd.DataFrame({'x':[1,2,3,4,5]})
df
```

```python
df['x'].min()
```

```python
df['x'].max()
```

Range = maximum value - minimum value 

```python
df['x'].max() - df['x'].min()
```

Deviation

```python
df['x-m'] = df['x']-df['x'].mean()
df
```

Mean Deviation

```python
df['x-m'].mean()    # mean deviation is always zero 
```

Population variance 

```python
df['x'].var(ddof=0)
```

Population Standard deviation 

```python
df['x'].std(ddof=0)
```

Sample variance 

df['x'].var(ddof=1)

Sample Std.deviation

```python
df['x'].std(ddof=1)
```

Coefficient of variance 

```python
df['x'].std(ddof=0)/df['x'].mean()
```

### Percentile

```python
df = pd.DataFrame({'x':[10,11,12,25,25,27,31,33,34,34,36,36,43,50,59]})
df
```

```python
# 0 percentile or minimum 
df['x'].quantile(0)
```

```python
# 25 percentile (Q1)
Q1 = df['x'].quantile(0.25)
Q1
```

```python
# 50 percentile (Q2)
Q2 = df['x'].quantile(0.5)
Q2
```

```python
# 75 percentile (Q3)
Q3 = df['x'].quantile(0.75)
Q3
```

```python
# 100 percentile or maximum
df['x'].quantile(1)
```

Inter Quartile Range(IQR) = Q3 - Q1

```python
IQR = Q3-Q1
IQR
```

Lower limit --> Q1 - (1.5*IQR)

```python
ll = Q1 - (1.5*IQR)
ll
```

Upper Limit --> Q3 + (1.5*IQR)

```python
ul=Q3 + (1.5*IQR)
ul
```
