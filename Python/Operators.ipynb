```python
x=5
print(x)
```

#### Logical operators / Boolean operators 

##### --> logical operators are used to check multiple condition at a time

##### --> when we have 'or' operator... atleast one condition should be true then only overall output will be true.
##### --> when we have 'and' operator... all conditions should be true then only overall output will be true.

#### Identity operators

##### -> identity operators are used to check of two values (or variables) are indicating to same object or not 
##### -> it returns output as boolean 
##### -> is, is not are the identity operators in the python 
##### -> is operator( is - true if the operens are identical 
##### -> is not operator (is not - true if the operends are not identical 

```python
a=5
b=5
print(a is b)   # id(a)==id(b)
```

```python
s1='Durga'
s2='durga'
print(s1 is s2)    # id(s1)==id(s2)
```

```python
a=6
b=8
print(a is not b)
```
