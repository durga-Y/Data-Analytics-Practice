write a program that gives sum of a even numbers in a list 

```python
l=[4,9,14,16,21,29,33]
even_list=[]
odd_list=[]
for i in l:
    if i%2==0:
        even_list.append(i)
    else:
        odd_list.append(i)
        
print(even_list,odd_list)
```

write a program that gives sum of even numbers and the sum of odd numbers 

```python
l=[4,9,14,16,21,29,33]
even_list=[]
odd_list=[]
odd_sum=0
even_sum=0
for i in l:
    if i%2==0:
        even_list.append(i)
        even_sum=even_sum+i
    else:
        odd_list.append(i)
        odd_sum=odd_sum+i

print(even_list, "and their sum",even_sum, odd_list,"and their sum",odd_sum)
        
          
        

    
```
