```python
#lst.copy() ----------> used to create a new list object with copy of items
```

```python
a=[1,2,3,4,5]
b=a.copy()
b.append(6)
print(b)
print(a)
```

```python
#Difference between shallow copy and deep copy  
```

```python
a=[1,2,3,4,5]          # Shallow copy -----> it indicates to different object 
b=a.copy()
b.append(6)
print(b)
print(a)                                       
```

```python
a=[1,2,3,4,5]          # deep copy -----> it indicates the same object 
b=a
b.append(6)
print(b)
print(a) 
```
