### 1 sample Z - Test

Case study 1 : The lengh of 12 sample of a fabric taken at random historic mean and standard deviation are 150 cm and 5.5 cm respectively, company manager wants to test if the current mean is equal to the historic mean assume alpha to be 0.05

### step 1 - formukate of H0 and H1

H0 : mean = 150 

Ha : mean != 150

### Step-2. Select level of significance 

alpha = 0.05

```python
df = pd.read_excel(r"D:\OneDrive\5.Data Analytics\Notes\Fabric data.xlsx")
```

```python
df.head()
```

### Step-3. Check for normality 

```python
df['Fabric_length'].skew()
```

(Or)

```python
import seaborn as sns
sns.kdeplot(df['Fabric_length'])
```

(or)

**Shapiro Test - (to check normal distribution or not )**

```python
from scipy import stats
stats.shapiro(df['Fabric_length'])
```

### Step-4. Select the statistical test and calculate the P- value  

1 sample Z - Test (Two tail Rejection)

```python
from statsmodels.stats.weightstats import ztest 
ztest = ztest(df['Fabric_length'], value = 150, alternative = 'two-sided')
ztest
```

### Step-5. Based on P-value accept or reject H0

* P < alpha
* P low -- null go
* Rejest H0 --> mean!=150(action required)
