```python
# option 2
```

```python
marks = {'history':45, 'Hindi':56}
marks.update({'English':47})
marks
```

```python
# To add sinle or multiple items to a dictionary
#dict.update()
```

```python
marks = {'history':45, 'Hindi':56}
marks.update({'chemistry':65,'physics':85})
marks
```

```python
# To delete an item in dictionary
# option 1----> del dict[key]  
```

```python
marks = {'history':45, 'Hindi':56}
del marks['Hindi']
marks
```

```python
# option2 ---> dict.pop(key)
# this method will remove and retirn the value of the specified key 
# if a key does not exist it raises a KeyError 
```

```python
marks = {'history':45, 'Hindi':56}
marks.pop('history')
marks
```

```python
# dict.clear() ---> methos removes all items from the dictionary 
```

```python
marks = {'history':45, 'Hindi':56}
marks.clear()
marks
```

```python
# dict.copy() ----> methos returns a shallow copy of a dictionary 
#changes to the copy of dictionary do not affect the original dictionary 
```

```python
d = {'history':45, 'Hindi':56}
a = d.copy()
a.update({'Maths':35})
print(a)
print(d)
```
