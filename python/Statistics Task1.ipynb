finding Measures of dispersion/spread (Range-(min(),Max, variance,std.deviation )

```python
df= pd.DataFrame({'salaries':[2000,2500,2700,2800,3000,3200,3500,4000,4200,4500,4700,5000,5200,5500,6000]})
print("Range =", df['salaries'].max()-df['salaries'].min())
print("Variance=", df['salaries'].var())
print('Std_deviation=', df['salaries'].std())
```

observing values of measures of dispersion with an outlier affect  

```python
df.iloc[len('salaries')] = 100
print("Range =", df['salaries'].max()-df['salaries'].min())
print("Variance=", df['salaries'].var())
print('Std_deviation=', df['salaries'].std())
```

by above observations outlier is more affected the dispersion values

```python
df = pd.DataFrame({'x':[10,20,14,16,18,20,22,25,30,35,40,45]})
df 
```

```python
print('min =', df['x'].quantile(0))
print('max =', df['x'].quantile(1))

```

```python
q1= df['x'].quantile(0.25) # 25%
q1
```

```python
q2=df['x'].quantile(0.50) # 50%
q2
```

```python
q3 =df['x'].quantile(0.75) # 75%
q3
```

find IQR and outlier 

```python
df = pd.DataFrame({'x':[10,12,14,16,18,20,22,25,30,35,40,50]})
df
```

```python
IQR = q3 -q1
```

```python
L_limit = q1 - (IQR*1.5)
U_limit = q3+ (IQR*1.5)
```

```python
print("IQR :", IQR)
print("L_limit:", L_limit)
print("U_limit:", U_limit)
```

```python
df[(df['x'] < L_limit) & (df['x'] > U_limit)]
```

```python
df = pd.DataFrame({'x':[4,8,6,5,3,8,9]})
M = df['x'].sum()/len(df)
x1 = df['x']-M
x2 = x1**2
var = x2.sum()/len(df)
std_dev = np.sqrt(var)
print("Mean : ", M)
print("var :", var)
print("std :", std_dev)
```
